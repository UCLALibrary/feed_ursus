```python
import re
from pathlib import Path

FIELD_NAME = re.compile(r"\b\w+_[tdsi]t?e?s?i?m?\b")
URSUS_DIR = "../../ursus"

EXCLUDE = {"node_modules", ".git", ".venv"}
FALSE_MATCHES = {"to_s"}

catalog_controller = Path(URSUS_DIR) / "app/controllers/catalog_controller.rb"

ursus_repo_fields = {"id"} | {
    match
    for match in FIELD_NAME.findall(catalog_controller.read_text())
    if match not in FALSE_MATCHES
}

print(sorted(ursus_repo_fields))
```

```python
from feed_ursus.solr_record import UrsusSolrRecord

solr_fields = {
    *UrsusSolrRecord.model_fields.keys(),
    *UrsusSolrRecord.model_computed_fields.keys(),
}
print(sorted(solr_fields))
```

```python
missing = ursus_repo_fields - solr_fields
print("\n".join(sorted(missing)))
```

```python
extra = solr_fields - ursus_repo_fields
print("\n".join(sorted(extra)))
```
